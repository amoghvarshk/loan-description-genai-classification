## Creating function to apply to larger amounts of data
* There was also some changes made to the 'prelude' due to errors found such empty strings returning full responses which took the form of loan descriptions not categories.

```python
def labelReview(desc):
    prelude = """
You are a helpful lending club employee attempting to discern what type of loan is most prevelent on the site.

Your objective is to identify loans as one of the following:

Business related such as to start or continue business operations,

Home improvement related for a renevation, add on or other improvements,

Other which includes anything that does not fall into the first two categories.

You should respond with  "Business", "Home", or "Other" with no additional information provided.

You should not misconstrude someone simply saying they are a good employee as meaning the loan is about business. The loan must be about improving or helping with business operations.

Below you can find some examples:

Input: Borrower added on 12/11/12 > To purchase a Pellet fired Stove for our Homestead, to reduce the Electric heat cost that we pay on a monthly basis.  Estimated Savings during heat season would be approximately $1200. annually<br>
Output: Home Improvement
Input: my business depends on good transportation and am currently in need of a new truck
Output: Business
Input:  Borrower added on 04/08/10 > This loan is for the purchase of a used motorcycle.
I opted to use Lending Club as an alternative to financing the purchase due the high interest rates associated with traditional lenders.
  I am very dependable when it comes to paying back anything.  I am the Chief Probation Officer for a rural county in northern
  California.  I have held the same job since I got out of the Marine Corps in 1994.<br/> Borrower added on 04/08/10 > In addition to my income,
  my wife is employed as an elementary school teacher for the last 7 years.  Both of our jobs are secure and combined we make approximately $152,000 per year.
  I am as reliable as they come and look forward to hopefully being funded.<br/> Borrower added on 04/08/10 > I plan on using the funds to purchase a used motorcycle - my first.
  I  opted to go with Lending Club because the interest rates are much more reasonable than that of a traditional bank despite my excellent credit rating.
  I am about as good as they come as it relates to loan repayments. I have always been very dependable/responsible with my money.
  In addition to my income, my works full time as well as an elementary school teacher and combined we make nearly $150,000 per year.  Our jobs are very secure.
  I am in law enforcement and am currrently the Chief of Probation in a rural county in northern California.  I have worked in the same department since I got out of the Marine Corps in 1994.
  Thank you for taking the time to read this - I look forward to being funded.<br/>
Output: Other

Please respond for the following review, indicating "Business", "Home", or "Other" regarding the main reason for the loan request.
These should be the only three responses.

If there is nothing below this line please respond with "Other" You should not give a response longer than 1 word.

"""
    query = prelude + desc
    response = client.models.generate_content(
        model='gemini-2.0-flash',
        contents=query
    )
    return response.text

labelReview(subset["desc"][0])
```

##5) Time and Cost Estimates
### Free Tier:
##### Cost:
* Luckily this part is easy it is free though this will lead to a much more time required to process this huge amount of data.
##### Time Required:
* At minimum this process would take 85 days. This daunting number comes from the fact that we would require a separate request for each of the 126,053 loan descriptions, and the free tier has a limit of 1,500 requests per day.

### Tier 1:
##### Cost:
* The cost of the paid tier would be roughly $4370.89. The bulk of this cost comes from the grounding used in Gemini and the cost per thousand requests after a certain limit. The calculation was done by dividing te number of tokens (roughly 7562400) dividing by 1000000 and multiplying across the costs per million tokens and per thousand requests. Calculation as follows:(input price).76 + (output price)3.03+(context caching).19 + (context caching storage)7.56+ (grounding)4359.355
##### Time Required:
* This is where the tier up comes with large perks. As there is no daily limit on requests in this tier, what we have to worry about is the requests per minute limit which is 2,000. This should allow the entire dataset to be completed in just under 1 hour and 4 minutes. The token per minute rate should not be an issue as there is an estimated total of under 8 million tokens for the entire dataset and the token per minute limit is 4 million.
